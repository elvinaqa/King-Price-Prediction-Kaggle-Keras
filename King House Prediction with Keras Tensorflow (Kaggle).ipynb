```python
# Feature Engineering
```

```python
df = df.drop('id', axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot('month', 'price', data=df)
```

```python
df = df.drop('date', axis=1)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
sns.scatterplot('zipcode', 'price', data=df)
```

```python
df = df.drop('zipcode', axis=1)
```

```python
X = df.drop('price', axis=1).values
```

```python
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, 
          validation_data=(X_test, y_test),
         batch_size=128,
         epochs=444)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, mean_absolute_error, explained_variance_score, balanced_accuracy_score 
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)**0.5
```

```python
mean_absolute_error(y_test, predictions)
```

```python
explained_variance_score(y_test, predictions)
```

```python
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'y')
```
